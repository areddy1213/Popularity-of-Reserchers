```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import math
from collections import Counter, OrderedDict, defaultdict
import numpy as np
```

```python
import seaborn as sns
```

```python
final_df = pd.read_csv('final_with_ranks.csv')
```

```python
final_df.columns
```

```python
final_df['centrality'].fillna(0,inplace = True)
```

```python
def normalize(dataset):
    dataNorm=((dataset-dataset.min())/(dataset.max()-dataset.min()))*100
    #dataNorm["Author"]=dataset["Author"]
    return dataNorm
```

```python
new_final = final_df[[ 'Cum_Page_Rank', 'num_citation', 'centrality','Katz_centrality','hindex',]]
```

```python
new_final = normalize(new_final)
```

```python
new_final['Author'] = final_df['Author']
```

```python
new_final['Cum_Page_Rank'] = new_final['Cum_Page_Rank']*(1.5)
new_final['hindex'] = new_final['hindex']*(0.9)
```

```python
new_final['Final_Rank'] = new_final['Cum_Page_Rank'] + new_final['num_citation'] + new_final['centrality']+new_final['Katz_centrality'] + new_final['hindex']
```

```python
new_final['Centrality_score'] =  new_final['centrality']+new_final['Katz_centrality'] 
```

```python
new_final["Final_S_Rank"] = new_final["Final_Rank"].rank(ascending= False)
```

```python
new_final["Metric_Rank"] = new_final["Cum_Page_Rank"].rank(ascending= False)
```

```python
new_final['h_index_rank'] = new_final["hindex"].rank(ascending= False)
```

```python
new_final['centrality_rank'] = new_final["centrality"].rank(ascending= False)
```

```python
new_final['Katz_centrality_rank'] = new_final["Katz_centrality"].rank(ascending= False)
```

```python
new_final['Centrality_rank'] = new_final["Centrality_score"].rank(ascending= False)
```

```python
new_corr = new_final[[ 'Cum_Page_Rank', 'num_citation', 'centrality','Katz_centrality','hindex','Final_Rank']]
```

```python
correlation = new_corr.corr(method='spearman')
columns = correlation.nlargest(6, 'Final_Rank').index

correlation_map = np.corrcoef(new_corr[columns].values.T)
sns.set(font_scale=1.0)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

heatmap = sns.heatmap(correlation_map, cbar=True, annot=True, square=True, fmt='.2f', yticklabels=columns.values, xticklabels=columns.values)

plt.show()
```
